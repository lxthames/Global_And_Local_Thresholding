```python
import matplotlib.pyplot as plt
import cv2
from skimage import data
from skimage.filters import threshold_otsu
image = data.page()  #for getting the test image

#For manual global thresholding
global_thresh = threshold_otsu (image)
binary_global = image > global_thresh

#for 1 line global thresholding
# ret, bw_img = cv2.threshold(image, 150, 250, cv2.THRESH_BINARY) #where bw_img is the global thresholded image
# plt.imshow(bw_img)
# plt.show

#local Thresholding
thresh2 = cv2.adaptiveThreshold(image, 255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,199,5) #199,5 is the limit of gaussian

#Just a method to print a number of images in a specific format
fig, axes = plt.subplots ( nrows=3, figsize= (7,8)) #rows is the number of images 

ax0,ax1,ax2=axes
plt.gray()

ax0.imshow(binary_global)
ax0.set_title('Global Thresholding')
ax1.imshow(thresh2)
ax1.set_title('Local Thresholding')
ax2.imshow(image)
ax2.set_title('Orignal Image')

plt.show()
```
